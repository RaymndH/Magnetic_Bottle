# Project Writeup
The focus of my project was mapping the confinement of a charged particle in a "magnetic bottle". Magnetic bottles, also known as magnetic mirrors, have been in use since the 1950's as a means to contain fast moving charged particles. 

The principle they work on is based on the Lorentz force, which imparts a tangential force on moving particles in a magnetic field. While in a uniform field, this force can do no work, in a "bottle" where the magnetic field lines are more concentrated at the top and bottom, this causes moving particles to gyrate around field lines, until their direction of movement is reversed. While this is not a perfect system and particles with incorrect velocities will escape, it provides an interesting look into the field of magnetic confinement systems.

# Particle mover
This is the chunk of code which is used to find the movement of the particle. It uses something known as the Boris algorithm, which was developed in the 1940's. It is similar to the leap-frog integrator used, but is not a symplectic algorithm due to the acceleration depending on velocity. However, it does have the property of preserving phase space and is a natural choice for many plasma simulation algorithms.

It shares the interesting property of moving the velocity back half a step and using the average velocity to adjust the position, also known as the drift-kick approach.



The simulation part of the code accepts a particle with initial position and velocity, as well as functions for calculating magnetic and electric fields, time step size, total time to run, and optional mass and charge values. The code will break if the particle moves more than about 20 units away from the center.

```python
#define the mover which will use the boris algorithm 
#code adapted from www.particleincell.com
m=1.0
q=1.0
dt = .01
tot_time = 10
def push(part, dt):
    part[0] += part[1]*dt

def Boris(part, E, B, dt):
    
    #calculate t
    t= q*m*dt/2*B
   
    #calculate s from t
    s= 2*t / (1+sum(t*t)**.5)
    #calculate various velocities
    v_minus = part[1] + q/(2*m)*E*dt
    v_prime = v_minus + np.cross(v_minus, t)
    v_plus = v_minus + np.cross(v_prime, s)
    #v n+1/2
    part[1] = v_plus + q/(2*m)*E*dt
    return


def simulation(part, Efunc, Bfunc, dt, tot_time, m=1, q=1):
    t= np.arange(0,tot_time,dt)
    t=[]
    posarr = np.array([])
    velarr = np.array([])
    #push velocity back half step
    E = Efunc(part[0])
    B = Bfunc(part[0])
    print(B)
    Boris(p1, E, B, -.5*dt)
    i=0
    while i < tot_time/dt: 
        E = Efunc(part[0])
        B = Bfunc(part[0])
        Boris(part, E, B, dt)
        posarr = np.append(posarr, part[0])
        velarr = np.append(velarr, part[1])
        push(part, dt)
        if i>0:
            if i*dt*20/tot_time%1==0:
                print('timestep ',i*dt," of ", tot_time)
                print(part[0])
        i+=1
        t.append(i*dt)
        if sum(part[0]**2)**.5>20:
            print('simulation broke at time ',i*dt)
            break
    posarr = np.reshape(posarr, [len(t),3])
    velarr = np.reshape(velarr, [len(t),3])
    #return positions and velocities
    return (posarr, velarr)


```

## "Magnetic Bottle"
The code below simulates a magnetic bottle as 2 (very strong!) magnetic dipoles placed apart. A more accurate simulation would construct a ring of charge and then integrate using the bio-savart law, but this should be a fair approximation as long as we don't get too close to the dipoles.

```python
#now try with analytic solution for helmholtz coils (that is, 2 magnetic dipoles)
#field assumed to be constant in time
#code inspired by https://uio-ccse.github.io/computational-essay-showroom/essays/exercises/essayseeds/MagneticBottle_Numpy.html
mu_0 = 1
dipole_distance = 10
mu = np.array([0,0,5000])
def B_bottle(pos):
    x,y,z = pos;
    #set the position of the top dipole r prime
    rp_top = (0.0, 0.0, dipole_distance)
    #calculate r vector
    r_top = np.array([x,y,z]) - rp_top
    #print(r_top)
    #find magnitude of r
    rmag = sum(r_top**2)**.5
    #print(rmag)
    #B(r)
    Btop = mu_0/(4*np.pi)*(3*r_top * np.dot(mu, r_top)/np.power(rmag,5) - mu/np.power(rmag,3))
    #repeat for bottom dipole
    rp_bot = (0.0, 0.0, -dipole_distance)
    #calculate r vector
    r_bot = np.array([x,y,z]) - rp_bot
    #find magnitude of r
    rmag = sum(r_bot**2)**.5
    #B(r)
    Bbot = mu_0/(4*np.pi)*(3*r_bot * np.dot(mu, r_bot)/rmag**5 - mu/rmag**3)
    return (Btop + Bbot)
```
